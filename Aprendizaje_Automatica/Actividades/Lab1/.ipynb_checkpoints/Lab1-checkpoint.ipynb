```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.model_selection import GridSearchCV
```

```python
#La fuente de datos proporcionada es https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/
#Utilizamos los datos del fichero .data, y asignamos los nombres de las columnas descritos en .names
data = pd.read_csv("auto-mpg.data",delim_whitespace=True,names=['mpg',
'cylinders',
'displacement',
'horsepower',
'weight',
'acceleration',
'model year',
'origin',
'car name'])

data.head()
```

```python
#Definimos qué valores del dataset son las características y cuál el objetivo
caracteristica = ["cylinders",
"displacement",
"horsepower",
"weight",
"acceleration",
"model year",
"origin",
"car name"]
objetivo = "mpg"
```

```python
from sklearn.preprocessing import LabelEncoder
data['car name'] = LabelEncoder().fit_transform(data['car name'])
data=data.replace('?', np.nan).ffill()

car_train, car_test, tar_train, tar_test = train_test_split(data[caracteristica],
data[objetivo],test_size = 0.25,random_state=2) 

car_train.head()
```

```python
#Para buscar la configuración más óptima, 
#Definimos dos posibles tipos de kernel. Un kernel radial y otro lineal
kernels=['rbf','linear']
#Definimos varios valores de C, que determinan el tamaño del softmargin, para probar cuál de ellos es más óptimo
cs=[0.001, 0.10, 0.1, 1, 10]
#Definimos varios valores de gamma, que determinan el radio de influencia de cada punto
gammas=[0.001, 0.01, 0.1, 1]

#Inicializamos las variables que guardarán los mejores valores encontrados
maxscore=-100
maxkernel=''
maxcs=0
maxgamma=0
for i in kernels:
    for j in cs:
        for k in gammas:

            model=svm.SVR(kernel=i,C=j, gamma=k)
            model.fit(car_train, tar_train) 

            pred = model.predict(car_test)
            scor=model.score(car_test, tar_test)
            if(scor>maxscore):
                maxscore=scor
                maxkernel=i
                maxcs=j
                maxgamma=k
                
print("Se ha conseguido un score de "+
      str(maxscore)+" utilizando los parámetros: C: "+str(maxcs)+", Kernel: "+maxkernel+", Gamma: "+str(maxgamma)) 
            
```
