# Tarea 1
El objetivo de esta actividad es continuar familiarizándonos con el entorno Python. Para ello, crea un programa que automáticamente:

	1. Lee el fichero proporcionado.
	2. Cuenta el número de líneas del fichero.
	3. Cuenta el número de veces que aparece el artículo «el» en el fichero (nota: elemento contiene los caracteres «el» pero no debe contar puesto que no contiene a el artículo «el».
	4. Elimina las tildes del fichero y convierta todas las letras a minúscula.
	5. Crea un fichero de texto con el nombre «resultado.txt» que contendrá en la primera línea el número de líneas del fichero, en la segunda línea el número de veces que aparece el artículo «el», y a partir de la tercera línea el texto original todo en minúsculas y sin tilde.


```python
# Read the complete file
with open('./FicheroALeer.txt', 'r', encoding = "ISO-8859-1") as f:
    text = f.read()

# convert the text to lower case
text = str.lower(text)
```

```python
# Extract all new line symbols indices
nlines = []
for i, c in enumerate(text):
    if c == '\n':
        nlines.append(i)
        
# Print number of new lines
print(f'Number of lines = {len(nlines)}')
```

```python
# Find all accurances of 'el'
els = []
cpos = text.find('el ', 0)
while(cpos > 0):
    els.append(cpos)
    cpos = text.find('el ', cpos+1)
    
print(f'Number of uses of the article \'el\' = {len(els)}')
```

```python
# Translate acións to normal letters
simple_text = text.translate(str.maketrans('áéóíúñ', 'aeoiun'))
```

```python
# Write the results to a file
with open('./resultado.txt', 'w') as f:
    f.write(f'{len(nlines)}\n{len(els)}\n{simple_text}')

```

* Crea un DataFrame de Pandas donde la primera columna sean todas las provincias que aparecen en el archivo, la segunda columna el número total de habitantes por provincia, la tercera la desviación típica en el número total de habitantes por provincia y la cuarta el número total de municipios por provincia.

```python
pm = pd.read_csv('./poblacionMunicipios.csv')
pm.head()
```

```python
# Filter empty regions
pm2 = pm[pm['Poblacion'] > 0]

# Aggregate the population size per municipio
muni_pop = pm2[['Municipio', 'Poblacion']].groupby('Municipio').sum()
muni_pop.head()
```

```python
pm2['Municipio'].unique()
```

```python
# Crea un DataFrame de Pandas donde la primera columna sean todas las provincias que aparecen en el archivo,
# la segunda columna el número total de habitantes por provincia, la tercera la desviación típica en el número
# total de habitantes por provincia y la cuarta el número total de municipios por provincia.

# Aggregate and analyze by province
pm_analysis = pm2.groupby('Provincia').agg({'Poblacion':['sum', 'std'], 'Municipio': lambda x: len(x.unique())})
pm_analysis.head(10)
```

# Tarea 3
	• Carga el archivo «CP_Municipios.csv».
	• Empleando el código de provincia y el código de municipio, cruza los dos datasets comentados hasta el momento. El resultado final debe ser un DataFrame.
	• Genera un CSV llamado «faltan.csv» con todos los códigos postales que estén en «CP_Municipios.csv» pero que no tengan en «poblacionMunicipios.csv» ningún municipio con el que poder cruzar. Si todo cruza correctamente el archivo «faltan.csv» estará vacío. Se puede decir que «faltan.csv» son los códigos postales que se han perdido en el cruce a la hora de considerar la población.
	• Agrupa la información por código postal. Se quiere un DataFrame con las siguientes columnas: 
		○ Código postal.
		○ Número de municipios que tienen dicho código postal asignado.
		○ Población: se calculará como la suma de la población de todos los municipios que incluyen a dicho código postal.
		○ Provincia: a la que está asignada el código postal.

```python
# • Carga el archivo «CP_Municipios.csv».
cp = pd.read_csv('./CP_Municipios.csv')
cp.head()
```

```python
# • Empleando el código de provincia y el código de municipio, cruza los dos datasets comentados hasta el momento. El resultado final debe ser un DataFrame.
complete = pd.merge(pm2, cp, on=['CodProvincia', 'CodMunicipio', 'Municipio'])
complete.head(20)
```

```python
# • Genera un CSV llamado «faltan.csv» con todos los códigos postales que estén en «CP_Municipios.csv» pero
# que no tengan en «poblacionMunicipios.csv» ningún municipio con el que poder cruzar.
# Si todo cruza correctamente el archivo «faltan.csv» estará vacío. Se puede decir que «faltan.csv» son los
# códigos postales que se han perdido en el cruce a la hora de considerar la población.
faltan = set(cp['CP'].drop_duplicates().values) - set(complete['CP'].drop_duplicates().values)
faltan_df = pd.DataFrame(data={'CP': list(faltan)})
faltan_df.to_csv('./faltan.csv', index=False)
faltan_df.head(10)
```

```python
# Agrupa la información por código postal. Se quiere un DataFrame con las siguientes columnas: 
#    ○ Código postal.
#    ○ Número de municipios que tienen dicho código postal asignado.
#    ○ Población: se calculará como la suma de la población de todos los municipios que incluyen a dicho código postal.
#    ○ Provincia: a la que está asignada el código postal.
        
cp_analysis = complete.groupby('CP').agg({'Municipio': lambda x: len(x.unique()), 'Poblacion': 'sum', 'Provincia': 'first'})
cp_analysis.reset_index(inplace=True)
cp_analysis.head(20)
```
